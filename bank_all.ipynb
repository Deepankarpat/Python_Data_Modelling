```python
import pandas as pd
```

```python
df = pd.read_csv('bank-full.csv', sep=';')
```

```python
df.head()
```

```python
df.keys()
```

```python
df.shape
```

```python
df.dtypes
```

```python
p = df.drop(['age','balance','day','duration','campaign','pdays','previous'], axis=1)
```

```python
p.head()
```

```python
y = pd.get_dummies(p,columns=['job','marital','education','default','housing','loan','contact','month','poutcome','y'],drop_first=True)
```

```python
y.head()
```

```python
df.head()
```

```python
q = df.drop(['job','marital','education','default','housing','loan','contact','month','poutcome','y'],axis=1)
```

```python
q.head()
```

```python
r= pd.concat([y,q],axis=1)
```

```python
r.head()
```

```python
r.shape
```

```python
target = r['y_yes']
```

```python
X= r.drop(['y_yes'],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,target,test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,predictions)
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
